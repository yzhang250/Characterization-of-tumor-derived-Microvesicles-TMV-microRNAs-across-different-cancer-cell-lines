```python
import pandas as pd
from selenium import webdriver
import urllib
import matplotlib.pyplot as plt
```

```python
CELL_TYPE = "LOX"
up_regulated_abundant_miRs = pd.read_csv("./{} up_regulated_abundant_miRs.csv".format(CELL_TYPE), index_col=[0])
```

```python
url1HTTP='http://snf-515788.vm.okeanos.grnet.gr/#mirnas='
url3Middle='&methods='
url5Last='&selection=2'

# Generate a url for mirPath with a miRList and method could be targetScan or microT-CDS
def generateUrl(miRNA_list, method):
    url2MiR=";".join(miRNA_list)
    url4Method=";".join([method]*len(miRNA_list))

    url = url1HTTP+url2MiR+url3Middle+url4Method+url5Last
    print("Use this url to check the pathways affectd by the miRNAs: \n"+url)
    return url


miRNA_list = up_regulated_abundant_miRs["Unnamed: 0.1"]

url = generateUrl(miRNA_list, "microT-CDS")
```

```python
import time
from selenium import webdriver

url_heatmap = "http://snf-515788.vm.okeanos.grnet.gr/index.php?r=mirpath/heatmapViewer"
driver = webdriver.Chrome(executable_path='/usr/local/bin/chromedriver')  # Optional argument, if not specified will search path.
driver.get(url);
time.sleep(30) # Let the user actually see something!
button = driver.find_elements_by_xpath("//*[@id=\"heatmap\"]")[0]
button.click()
time.sleep(10) # Let the user actually see something!
next_window = driver.window_handles[1]
driver.switch_to.window(next_window)
download = driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr[2]/td/div/table/tbody/tr[1]/td[3]/a")[0]
download.click()
time.sleep(5)
next_window = driver.window_handles[2]
driver.switch_to.window(next_window)

file_name = './TMV_miRNA_pathway_enrichment_analysis/{} TMV upregulated miRNA involoved pathways.png'.format(CELL_TYPE)
f = open(file_name,'wb')
f.write(urllib.urlopen(driver.current_url).read())
f.close()
driver.quit()


print "Done with click, saved the miRNA pathway enrichment analysis heatmap file"
```

```python
img = plt.imread(file_name)

fig, ax = plt.subplots(figsize=(20,16))
ax.imshow(img)
```
