```python
import pandas as pd
samples = ["PcelA", "PcelB", "PcelC", "PMVA", "PMVB", "PMVC"]
series = []
for sample in samples:
    series.append(pd.read_table("Raw_counts/"+ sample +"_raw_count.txt", header=None, index_col=0, names = [sample]))
```

```python
for i in range(1, len(series)):
    series[0] = pd.concat([series[0], series[i]], axis=1)

df = series[0]
```
